```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family='AppleGothic')

path1 = './날짜별 대여량.csv'
df = pd.read_csv(path1, encoding='euc-kr',thousands = ',')
```

```python
df = df.drop( ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5'],axis=1)
df.dropna(axis=0,inplace=True)
```

```python
df['대여건수'] = df['대여건수'].apply(pd.to_numeric)
df['대여일시'] = pd.to_datetime(df['대여일시'],format='%Y-%m-%d')
df.dtypes
```

```python
df['month'] = df['대여일시'].dt.month
df['요일'] = df['대여일시'].dt.dayofweek
```

```python
df_month_mean = df.groupby('month').mean().drop('요일',axis=1)
df_month_mean
```

```python
plt.figure(figsize=(20,20))
df_month_mean.plot(kind='bar') ## 월별 분석
plt.title("달별 이용량",fontsize=25)
plt.xticks(rotation = 0,fontsize=15)
plt.show()
```

```python
df_day_of_week = df.groupby('요일').mean().drop('month',axis=1)
df_day_of_week
```

```python
plt.figure(figsize=(20,20))
df_day_of_week.plot() ## 월별 분석

plt.title("요일별 이용량",fontsize=25)
plt.show()
```

```python
for i in range(2,7):
    mask_month = df.month == i
    df_month = df[mask_month]
    d = df_month.groupby('요일').mean().drop('month',axis=1)
    d.plot()
```
